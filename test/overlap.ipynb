## Overlap

### 绿色mask(Ground Truth)

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

# Load the image
img_path = '/home/ec2-user/SeqX2Y_PyTorch/dataset/Overlap/P1/1_16/C0003.png'
img = Image.open(img_path).convert('L')  # Convert to grayscale

# Create a mask where the grayscale values are converted to green
img_array = np.array(img)
green_mask = np.zeros((img_array.shape[0], img_array.shape[1], 3), dtype=np.uint8)
green_mask[:, :, 1] = img_array  # Set the green channel to the image values

# Convert back to image
green_mask_img = Image.fromarray(green_mask)

# Save the green mask image
output_path = '/home/ec2-user/SeqX2Y_PyTorch/dataset/Overlap/P1/1_16/Green_Mask_C0003.png'
green_mask_img.save(output_path)
output_path

plt.figure(figsize=(6, 6))
plt.imshow(green_mask_img)
plt.title('Overlap of Ground Truth')
plt.axis('off')
plt.show()
```

### 紫色mask(Pred)

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

# Load the image
img_path = '/home/ec2-user/SeqX2Y_PyTorch/dataset/Overlap/P1/2D_16/C0003.png'
img = Image.open(img_path).convert('L')  # Convert to grayscale

# Create a mask where the grayscale values are converted to purple
img_array = np.array(img)
purple_mask = np.zeros((img_array.shape[0], img_array.shape[1], 3), dtype=np.uint8)
purple_mask[:, :, 0] = img_array  # Set the red channel
purple_mask[:, :, 2] = img_array  # Set the blue channel

# Convert back to image
purple_mask_img = Image.fromarray(purple_mask)

# Save the green mask image
output_path = '/home/ec2-user/SeqX2Y_PyTorch/dataset/Overlap/P1/2D_16/Purple_Mask_C0003.png'
purple_mask_img.save(output_path)

plt.figure(figsize=(6, 6))
plt.imshow(purple_mask_img)
plt.title('Overlap of Prediction')
plt.axis('off')
plt.show()
```

### 紫色 绿色 重叠

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from PIL import ImageEnhance

# Load the images again
predict_image = '/home/ec2-user/SeqX2Y_PyTorch/dataset/Overlap/P5/7_16/C0003.png' # 15_16
groundtruth_img = '/home/ec2-user/SeqX2Y_PyTorch/dataset/Overlap/P5/Ground_Truth/C0003.png' # P1
img1 = Image.open(predict_image).convert('L')
img2 = Image.open(groundtruth_img).convert('L')

# Create purple mask for the first image
img1_array = np.array(img1)
img1_purple_mask = np.zeros((img1_array.shape[0], img1_array.shape[1], 3), dtype=np.uint8)
img1_purple_mask[:, :, 0] = img1_array  # Red channel
img1_purple_mask[:, :, 2] = img1_array  # Blue channel

# Create green mask for the second image
img2_array = np.array(img2)
img2_green_mask = np.zeros((img2_array.shape[0], img2_array.shape[1], 3), dtype=np.uint8)
img2_green_mask[:, :, 1] = img2_array  # Green channel

# Convert masks back to images
img1_purple_mask_img = Image.fromarray(img1_purple_mask)
img2_green_mask_img = Image.fromarray(img2_green_mask)

# Blend the images with adjusted opacity to aim for gray in overlap
blended_img = Image.blend(img1_purple_mask_img, img2_green_mask_img, alpha=0.5)

# Enhance the brightness
enhancer = ImageEnhance.Brightness(blended_img)
brightened_img = enhancer.enhance(3)  # Increase brightness by 50%

plt.figure(figsize=(6, 6))
plt.imshow(brightened_img)

# Save the blended image
blended_output_path = '/home/ec2-user/SeqX2Y_PyTorch/dataset/Overlap/P5/7_16/overlap/GT_C0003.png'
brightened_img.save(blended_output_path)
blended_output_path
```
